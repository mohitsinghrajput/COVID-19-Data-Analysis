# COVID-19 Data Analysis

Implementation of basic concepts of Data Anaysis with the help of COVID-19 Data.

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel("Covid_cases_in_India.xlsx")
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
df
```

```python
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
df
```

```python
total_cases_overall=df['Total Cases'].sum()
print('The total number of cases till now in India is ',total_cases_overall)
```

```python
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
df
```

```python
df.style.background_gradient(cmap='Reds')
```

```python
Total_Active_Cases=df.groupby('Name of State / UT')['Total Cases'].sum().sort_values(ascending=False).to_frame()
Total_Active_Cases
```

```python
Total_Active_Cases.style.background_gradient(cmap='Reds')
```

### Graphical Represantaion

```python
#Bar plot for Indian spread of covid in different State / UT
df.plot(kind='bar',x='Name of State / UT',y='Total Cases',figsize=(15,8))
plt.xlabel('Name of State / UT')
plt.ylabel('Total Cases')
```

```python
#Scatter plot for Indian spread of covid in different State / UT
from matplotlib import style
style.use('ggplot')
df.plot(kind='scatter',x='Name of State / UT',y='Total Cases',figsize=(15,8),rot=90)
```

```python
#Reading data for different countries
dbd_India=pd.read_excel("per_day_cases.xlsx",parse_dates=True,sheet_name="India")
dbd_Italy=pd.read_excel("per_day_cases.xlsx",parse_dates=True,sheet_name="Italy")
dbd_Korea=pd.read_excel("per_day_cases.xlsx",parse_dates=True,sheet_name="Korea")
dbd_Wuhan=pd.read_excel("per_day_cases.xlsx",parse_dates=True,sheet_name="Wuhan")
dbd_India
```

```python
#Differen plots for Indian Dataset
fig=plt.figure(figsize=(13,6),dpi=80)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(dbd_India["Date"],dbd_India["Total Cases"],color='red')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in India")
fig=plt.figure(figsize=(13,6),dpi=100)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.bar(dbd_India["Date"],dbd_India["Total Cases"],color='blue')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in India")
fig=plt.figure(figsize=(13,6),dpi=150)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.scatter(dbd_India["Date"],dbd_India["Total Cases"],color='green')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in India")
```

```python
fig=plt.figure(figsize=(10,5),dpi=150)
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(dbd_India["Date"],dbd_India["Total Cases"],color='black',marker='*')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in India")
```

```python
#Plots for different country
fig=plt.figure(figsize=(10,6))
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(dbd_Italy["Date"],dbd_Italy["Total Cases"],color='red')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in Italy")
fig=plt.figure(figsize=(10,6))
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(dbd_Korea["Date"],dbd_Korea["Total Cases"],color='red')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in Korea")
fig=plt.figure(figsize=(10,6))
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(dbd_Wuhan["Date"],dbd_Wuhan["Total Cases"],color='red')
axes.set_xlabel("Date")
axes.set_ylabel("Total Cases")
axes.set_title("Confirmed cases in Wuhan")
```
